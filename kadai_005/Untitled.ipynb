```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2, 3, tight_layout=True)

y0 = [1, 2, -5, 2]
axes[0,0].plot(y0)
axes[0,0].set_title('折れ線グラフ')

x = np.linspace(0, 10, 100)
y1 = 2 + 2 * np.sin(2 * x)
axes[0,1].plot(x, y1)
axes[0,1].set_title('sin関数')

x2 = np.random.normal(15, 5, 2000)
axes[0,2].hist(x2, bins=30)
axes[0,2].set_title('ヒストグラム')

np.random.seed(3)
x3 = 4 + np.random.normal(0, 2, 60)
y3 = 4 + np.random.normal(0, 2, len(x3))
axes[1,0].scatter(x3, y3)
axes[1,0].set_title('散布図')

plt.show()
```
